```python
from sklearn.datasets import load_iris
iris=load_iris()
```

This data sets consists of 3 different types of irises' (Setosa, Versicolour and Virginica) petal and sepal length, stored in a 150x4 numpy.ndarray.

The rows being the samples and the columns being: Sepal Length, Sepal Width, Petal Length and Sepal Width.

```python
iris
```

```python
X = iris.data[:, (2, 3)] # petal length, petal width
```

```python
iris.target
```

### **Categories**

*   0 Setosa
*   1 Versicolour
*   2 Virginica

```python
y = (iris.target == 0)
```

```python
y
```

```python
y = (iris.target == 0).astype(np.int)
```

```python
y
```

```python
X
```

```python
from sklearn.linear_model import Perceptron
```

Documentation: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Perceptron.html

The multi-level Perceptron model can only be created using Keras, TensorFlow. But for single Perceptron model is created by sklearn.

```python
per_clf = Perceptron(random_state=42)
per_clf.fit(X, y)
```

```python
y_pred = per_clf.predict(X)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y, y_pred)
```

```python
per_clf.coef_
```
